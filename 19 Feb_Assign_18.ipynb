Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for Python, designed to be simple and lightweight while providing the necessary tools and features for building web applications. It is based on the Werkzeug WSGI toolkit and the Jinja templating engine.

Advantages of Flask Framework:

1. Lightweight and Minimalistic: Flask follows a minimalist approach, providing only the essential components for web development. It doesn't impose strict project structures or dependencies, allowing developers to have more flexibility and control over their application's design.

2. Easy to Get Started: Flask has a simple and intuitive API, making it beginner-friendly and easy to learn. Its documentation is comprehensive and well-documented, providing clear guidance for developers. Getting started with Flask requires minimal setup and boilerplate code.

3. Flexibility: Flask allows developers to choose and integrate the tools and libraries they prefer for various functionalities. It doesn't enforce specific databases or template engines, enabling developers to work with their preferred technologies.

4. Extensibility: Flask supports the use of extensions, which are additional libraries that provide extra functionality. These extensions cover a wide range of features such as database integration, form handling, authentication, and more. Flask's modular design and extensibility make it easy to integrate these extensions into the application.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs to specific functions or views within a web application. It is handled by the @app.route() decorator, which allows you to define the routes and associate them with the functions that will handle the corresponding requests.

App routes are used for several reasons:

1. URL Mapping: App routes allow you to define the URLs for different pages or functionalities within your application. Each route corresponds to a specific URL path, enabling users to access different parts of your application by navigating to the respective URLs.

2. Request Handling: By associating routes with specific functions, Flask can determine which function should handle incoming requests for a particular URL. When a user visits a specific URL, Flask automatically invokes the associated function to generate the appropriate response.

3. Modular Development: App routing in Flask promotes modular development by allowing you to organize your code into smaller, manageable functions. Each function can handle a specific route or page, making it easier to maintain and extend your application.
